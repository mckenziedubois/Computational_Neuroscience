```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
```

```python
tspan = np.linspace(0,10,101)
v_reset = 0
v_thresh = 1
init_cond = 0
istim = 1.7
b = -0.5
```

```python
dt = 0.1
num_times = len(tspan)
v_values = np.zeros(num_times)
Vmaxes = np.zeros(num_times)
v_values[0] = init_cond
I= 1.7
```

```python
dv_to_plot = np.zeros(num_times)
v_to_plot = np.zeros(num_times)

v_to_plot[0] = init_cond
for i in range(num_times-1):
    dv_to_plot[i] = istim + b - v_to_plot[i]
    v_to_plot[i+1] = v_to_plot[i] + (dv_to_plot[i]*dt)
```

```python
py.plot(v_to_plot,dv_to_plot)
py.title('dV versus V')
py.xlabel('V')
py.ylabel('dV')

#PHASE LINE: our system does not have an equilibrium values, and our dV is positive for all V values 
```

```python
for i in range((num_times)-1):
    dV = I + b - v_values[i]
    v_values[i+1] = v_values[i] + dt*dV
    if v_values[i+1] >= v_thresh:
        v_values[i+1] = v_reset
```

```python
py.plot(tspan,v_values)
py.xlabel('Time')
py.ylabel('Voltage')
py.xlim([0,tspan[-1]])
py.title('Solution Curve')
```
